### Task 6 : Prediction Using Decision Tree Algorithm

#### Import all necessary libraries

```python
# importing libraries
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split 
from sklearn.tree import DecisionTreeClassifier 
from sklearn.metrics import accuracy_score 
from sklearn import tree
```

```python
# loading the dataset 
df = pd.read_csv('dataset_5.csv') 
df.head()
```

```python
# checking the shape 
df.shape
```

```python
# checking for missing values, size, datatypes etc 
df.info()
```

```python
# getting a statistical summary 
df.describe()
```

```python
# creating the independent (X) and dependent (y) features 
X = df.drop(columns=['Species']) 
y = df['Species']
```

#### Splitting the data into train and test

```python
# splitting the data into training and testing sets 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=20)
```

```python
# training a DT model 
dt = DecisionTreeClassifier() 
dt.fit(X_train,y_train)
```

```python
# making predictions 
y_pred = dt.predict(X_test)
```

#### Caculating the accuracy of model

```python
# calculating the accuracy of the model 
print('Accuracy: ', (accuracy_score(y_test, y_pred) * 100))
```

```python
# plotting the decision tree 
plt.figure(figsize=(7,7)) 
tree.plot_tree(dt, feature_names=df.columns, class_names=['Setosa', 'Versicolor', 'Virginica'], filled = True, rounded=True);
```

#### Making Predictions with Trained Decision Tree Model on New Data.


```python
# Assuming 'new_data' is a DataFrame with the same features as your training data
new_data = pd.DataFrame({
    'SepalLengthCm': [5.1],
    'SepalWidthCm': [3.5],
    'PetalLengthCm': [1.4],
    'PetalWidthCm': [0.2],
})

# Use the trained model to predict on the new data
new_predictions = dt.predict(new_data)

# 'new_predictions' will contain the predicted class labels for the new data
print(new_predictions)
```
